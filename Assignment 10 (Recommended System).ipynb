```python
# converting long data into wide data using pivot table
data2= data.pivot_table(index='ID',columns='Book_Title',values='Book_Rating').reset_index(drop=True)
```

```python
data2
```

```python
#impute those NANs values with 0 values
data2.fillna(0,inplace=True)
data2
```

```python
# calculating cosine similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine ,correlation
```

```python
# Calculating Cosine Similarity between Users on array data
user_sim=1-pairwise_distances(data2.values,metric='cosine')
user_sim
```

```python
# Store the results in a dataframe format
user_sim2=pd.DataFrame(user_sim)
user_sim2
```

```python
# Set the index and column names to user ids 
user_sim2.index=data['ID'].unique()
user_sim2.columns=data['ID'].unique()
user_sim2
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,1)
user_sim2
```

```python
# most similar users
user_sim2.idxmax(axis=1)
```
